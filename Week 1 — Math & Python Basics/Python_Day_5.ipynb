1. Plot two lines on the same plot with different colors and labels.



````python
# Data
a = [1, 2, 3, 4, 5]
b = [10, 30, 50, 70, 90]
c = [5, 15, 25, 35, 45]  # Second line

# Create figure
plt.figure(figsize=(4, 3))

# Plot first line
plt.plot(a, b, color='blue', marker='o', label='Line 1')

# Plot second line
plt.plot(a, c, color='red', marker='s', label='Line 2')

# Add title and labels
plt.title("Two Lines Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Add legend
plt.legend()

# Show plot
plt.show()
````

2. Use subplots to display a line plot and a scatter plot side by side.



````python
# Sample data
x = [1, 2, 3, 4, 5]
y_line = [1, 4, 9, 16, 25]
y_scatter = [5, 15, 25, 35, 45]

# Create subplots (1 row, 2 columns)
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# Line plot on the first subplot
axs[0].plot(x, y_line, color='blue', marker='o')
axs[0].set_title("Line Plot")
axs[0].set_xlabel("X-axis")
axs[0].set_ylabel("Y-axis")
axs[0].grid(True)

# Scatter plot on the second subplot
axs[1].scatter(x, y_scatter, color='red')
axs[1].set_title("Scatter Plot")
axs[1].set_xlabel("X-axis")
axs[1].set_ylabel("Y-axis")
axs[1].grid(True)

# Adjust layout and show
plt.tight_layout()
plt.show()
````

6. Create a Seaborn histogram with kde=True for a random dataset of 50 numbers.



````python
import numpy as np

# Generate random dataset of 50 numbers
data = np.random.randn(50)

# Plot histogram with KDE
plt.figure(figsize=(6,4))
sns.histplot(data, kde=True, bins=10, color="skyblue", edgecolor="black")

plt.title("Seaborn Histogram with KDE")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
````

7. Customize Seaborn plots with style (sns.set_style) and palette.



````python
# Generate random dataset
data = np.random.randn(100)

# Set style and color palette
sns.set_style("whitegrid")   # Options: "white", "dark", "whitegrid", "darkgrid", "ticks"
sns.set_palette("husl")      # Options: "deep", "muted", "bright", "pastel", "dark", "colorblind", or custom

# Plot histogram with KDE
plt.figure(figsize=(6,4))
sns.histplot(data, kde=True, bins=15)

plt.title("Customized Seaborn Plot", fontsize=14)
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
````

1. Using Seaborn, plot a scatter plot with regression line (sns.regplot) for a dataset.



````python
# Sample dataset
x = np.arange(1, 21)
y = 3 * x + np.random.randn(20) * 5   # Linear relation with noise

# Plot regression line with scatter
plt.figure(figsize=(6,4))
sns.regplot(x=x, y=y, color="blue", marker="o")

plt.title("Scatter Plot with Regression Line", fontsize=14)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
````

2. Create a grouped bar chart using Matplotlib for multiple categories.



````python
# Sample data
categories = ['A', 'B', 'C', 'D']
group1 = [10, 20, 15, 25]
group2 = [12, 18, 22, 28]

x = np.arange(len(categories))  # category positions
width = 0.35  # width of each bar

# Plot grouped bars
plt.bar(x - width/2, group1, width, label='Group 1', color='skyblue')
plt.bar(x + width/2, group2, width, label='Group 2', color='orange')

# Customizations
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Grouped Bar Chart")
plt.xticks(x, categories)
plt.legend()

plt.show()
````

3. Generate a heatmap using Seaborn for a 5x5 correlation matrix.



````python
import pandas as pd

# Generate random data for 5 variables
np.random.seed(42)
data = np.random.rand(100, 5)  # 100 samples, 5 variables
df = pd.DataFrame(data, columns=['A', 'B', 'C', 'D', 'E'])

# Compute correlation matrix
corr_matrix = df.corr()

# Plot heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, linewidths=0.5)
plt.title("Correlation Heatmap (5x5)", fontsize=14)
plt.show()
````

4. Plot a stacked bar chart for two sets of sales data over 4 products.


````python
# Data
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales1 = [30, 25, 35, 20]   # Sales dataset 1
sales2 = [15, 20, 25, 30]   # Sales dataset 2

# Create stacked bar chart
x = np.arange(len(products))  # X positions

plt.bar(x, sales1, label='Sales 2024')
plt.bar(x, sales2, bottom=sales1, label='Sales 2025')

# Customization
plt.xticks(x, products)
plt.xlabel("Products")
plt.ylabel("Sales")
plt.title("Stacked Bar Chart of Sales Data")
plt.legend()

plt.show()
````

5. Customize tick labels, legends, and grid in Matplotlib.



````python
# Data
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]

plt.figure(figsize=(6, 4))

# Plot two lines
plt.plot(x, y1, label="Line A", color="blue", linestyle="--", marker="o")
plt.plot(x, y2, label="Line B", color="red", linestyle="-.", marker="s")

# âœ… Customize tick labels
plt.xticks(ticks=x, labels=[f"Point {i}" for i in x], rotation=30)
plt.yticks([0, 2, 4, 6, 8, 10])

# âœ… Add legend (custom position & style)
plt.legend(loc="upper left", fontsize=10, frameon=True, shadow=True)

# âœ… Add grid (customized)
plt.grid(True, which="both", linestyle=":", linewidth=0.7, color="gray")

# âœ… Labels and title
plt.xlabel("Custom X-axis")
plt.ylabel("Custom Y-axis")
plt.title("Customized Plot in Matplotlib")

plt.show()
````

6. Create a pairplot using Seaborn for a DataFrame with columns Age, Salary, Department.



````python
# Sample Data
data = {
    'Age': [25, 30, 35, 40, 45, 50],
    'Salary': [40000, 50000, 60000, 70000, 80000, 90000],
    'Department': ['HR', 'IT', 'Finance', 'IT', 'HR', 'Finance']
}

df = pd.DataFrame(data)

# âœ… Pairplot (hue = Department for colors)
sns.pairplot(df, hue="Department", diag_kind="kde", palette="Set2")

plt.show()
````
