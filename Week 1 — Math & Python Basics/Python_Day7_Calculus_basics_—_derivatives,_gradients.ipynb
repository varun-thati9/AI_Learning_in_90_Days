```python

def numerical_gradient(f, x, y, h=1e-5):
    dfdx = (f(x+h,y)-f(x-h,y))/(2*h)
    dfdy = (f(x,y+h)-f(x,y-h))/(2*h)
    return np.array([dfdx, dfdy])

f = lambda x,y: x**2 + y**2
numerical_gradient(f, 2, 3)
```

6. Visualize gradient descent steps on
ùëì
(
ùë•
)
=
ùë•
2
f(x)=x
2
.


```python
import numpy as np
import matplotlib.pyplot as plt

# Function and gradient
def f(x): return x**2
def grad(x): return 2*x

# Gradient Descent params
alpha = 0.2          # learning rate (0 < alpha < 1 for this quadratic)
x0 = -4.0            # starting point
steps = 10

# Run GD
xs = [x0]
for _ in range(steps):
    xs.append(xs[-1] - alpha * grad(xs[-1]))

# Plot function
xplot = np.linspace(-5, 5, 400)
plt.figure(figsize=(7, 4))
plt.plot(xplot, f(xplot), label="f(x) = x^2")

# Plot GD steps
ys = [f(x) for x in xs]
plt.scatter(xs, ys, zorder=3, label="GD steps")

# Arrows between steps
for i in range(len(xs)-1):
    plt.annotate("",
        xy=(xs[i+1], ys[i+1]),
        xytext=(xs[i], ys[i]),
        arrowprops=dict(arrowstyle="->", lw=1))

# Formatting
plt.title("Gradient Descent on f(x) = x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()

# Also print the iterates
for i, x in enumerate(xs):
    print(f"Iter {i:2d}: x = {x:.6f}, f(x) = {f(x):.6f}")
```
